```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
%matplotlib inline

from matplotlib import cm

road = cv2.imread('../DATA/road_image.jpg')
road_copy = np.copy(road)
road_copy2 = np.copy(road_copy)
plt.imshow(road)
```

```python
marker_image = np.zeros(road.shape[:2], dtype = np.int32)

segments = np.zeros(road.shape, dtype = np.uint8)
segments.shape
```

```python
def create_rgb(i):
    x = np.array(cm.tab10(i))[:3] * 255
    return tuple(x)
```

```python
colors = []

for i in range(10):
    colors.append(create_rgb(i))
    
n_markers = 10

current_marker = 1
marks_updated = False

def mouse_callback(event, x, y, flags, param):
    global marks_updated, road_copy2
        
    road_copy2 = road_copy.copy()
    
    cv2.circle(road_copy2, (x, y), 10, colors[current_marker], -1)
    
    if event == cv2.EVENT_LBUTTONDOWN:
        cv2.circle(marker_image, (x, y), 10, (current_marker), -1)
        
        cv2.circle(road_copy, (x, y), 10, colors[current_marker], -1)
        
        marks_updated = True
        
def mouse_callback_color(event, x, y, flags, param):
    global current_marker
    
    for i in range(n_markers):
        if event == cv2.EVENT_LBUTTONDOWN and i * 100 <= x and x < (i + 1) * 100:
            current_marker = i
        
def init_color_spectrum(image):
    for i in range(n_markers):
        cv2.rectangle(image, [i * 100, 0], [(i + 1) * 100, 100], tuple(map(int, colors[i])), -1)
```

```python
cv2.namedWindow('Road Image')
cv2.namedWindow('Color Spectrum')
cv2.setMouseCallback('Road Image', mouse_callback)
cv2.setMouseCallback('Color Spectrum', mouse_callback_color)

colorSpectrum = np.zeros([100, 1000, 3])

init_color_spectrum(colorSpectrum)

colorSpectrum = colorSpectrum.astype(np.uint8)

# print(colorSpectrum)

while True:
    cv2.imshow('WaterShed Segments', segments)
    cv2.imshow('Road Image', road_copy2)
    cv2.imshow('Color Spectrum', colorSpectrum)
    
    k = cv2.waitKey(1)
    
    if k == 27:
        break
        
    elif k == ord('c'):
        road_copy = road.copy()
        marker_image = np.zeros(road.shape[0:2], dtype = np.int32)
        segments = np.zeros(road.shape, dtype = np.uint8)
        
    elif k > 0 and chr(k).isdigit():
        current_marker = int(chr(k))
        
    if marks_updated:
        
        marker_image_copy = marker_image.copy()
        cv2.watershed(road, marker_image_copy)
        
        segments = np.zeros(road.shape, dtype = np.uint8)
        
        for color_ind in range(n_markers):
            segments[marker_image_copy == (color_ind)] = colors[color_ind]
            
        marks_updated = False
        
cv2.destroyAllWindows()
```
