```python
import torch
import torchvision.transforms as T
from torch import nn
import numpy as np
```

```python
transform = nn.Sequential(
    T.RandomRotation(30),
    T.RandomAffine(0, translate = [0.1, 0.1], shear = 0.2, scale = [0.8, 1.2]),
#     T.ToTensor(),
)
```

```python
s = np.transpose(np.transpose(transform(torch.Tensor(dog2)), axes = [2, 0, 1]).numpy(), [1, 2, 0])
plt.imshow(s)
```
